```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col = "coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width = 800,
    height = 400,
    rot = 90
)
```

---

### Prepare the Data

```python
# confirm column data type
df_market_data.dtypes
```

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data 
# Create a DataFrame with the scaled data
cols = df_market_data.columns
crypto_df_scaled = pd.DataFrame(StandardScaler().fit_transform(df_market_data[cols]), 
                                columns = cols)

# Copy the crypto names from the original data
# Set the coinid column as index
crypto_df_scaled.index = df_market_data.index

# Display sample data
crypto_df_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# function for generating df_elbow
def create_elbow(df):
    
    # lists of k-values (1 to 11) and inertia 
    k = list(range(1, 11))
    inertia = []
    
    # for loop to evaluate with K-means algorithm, fit model, append inertia
    for i in k:
        k_model = KMeans(n_clusters = i, random_state = 1)
        k_model.fit(df)
        inertia.append(k_model.inertia_)
        
    # DataFrame for k and inertia
    elbow_data = {'k': k, 'inertia': inertia}
    df_elbow = pd.DataFrame(elbow_data)
    return df_elbow
```

```python
# call function to create elbow df for scaled data
df_elbow = create_elbow(crypto_df_scaled)
df_elbow.head()
```

```python
# Plot elbow line chart to visually identify the optimal value for k.
elbow_plot = df_elbow.hvplot.line(
    x = "k", 
    y = "inertia", 
    title = "Elbow Curve", 
    xticks = "k" )
elbow_plot
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** I pick k=4 as the best value for elbow spot.

---

### Cluster Cryptocurrencies with K-means Using the Original Data

```python
# function to fit model, add cluster column

def fit_model(df, k):
    
    # Initialize the K-Means model using pre-determined k
    model = KMeans(n_clusters = k, random_state = 1)

    # Fit the K-Means model 
    model.fit(df)

    # Predict the clusters
    clusters = model.predict(df)

    # Create a copy of the DataFrame
    fit_df = df.copy()

    # Add a class column with the labels
    fit_df['Clusters'] = clusters

    return fit_df
```

```python
# call fit model function on scaled data
clusters_df = fit_model(crypto_df_scaled, 4)

# Display sample data
clusters_df.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cluster_plot = clusters_df.hvplot.scatter(x = "price_change_percentage_24h",
                           y = "price_change_percentage_7d",
                           by = 'Clusters',
                           hover_cols = 'coin_id',
                           title = 'Scatter Plot with Clusters, original features (k=4)')
cluster_plot

```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components = 3)
```

```python
# Use the PCA model with `fit_transform` to reduce to 3 principal components.
crypto_pca = pca.fit_transform(crypto_df_scaled)

# View the first five rows of the DataFrame. 
crypto_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information can be attributed to each principal component.
variance = pca.explained_variance_ratio_
print(variance)
print(f"{sum(variance)*100:.2f}%")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** about 89.5% of total variance are contributed by the 3 principal components.

```python
# Create a new DataFrame with the PCA data.
crypto_pca_df = pd.DataFrame(
    crypto_pca,
    columns = ['PCA1', 'PCA2', 'PCA3'])

# Copy the crypto names from the original data, set the coinid column as index
crypto_pca_df.index = df_market_data.index

# Display sample data
crypto_pca_df.head()
```

---

### Find the Best Value for k Using the PCA Data

```python
# call function to create elbow df for PCA data
df_elbow_pca = create_elbow(crypto_pca_df)
df_elbow_pca.head()
```

```python
# Plot a line chart with inertia against k to decide optimal k visually
elbow_plot_pca = df_elbow_pca.hvplot.line(
    x = "k", 
    y = "inertia",
    title = "Elbow Curve using PCA")

elbow_plot_pca
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** I pick k = 4 for the best value


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** best k value is same as using original data.

### Cluster Cryptocurrencies with K-means Using the PCA Data

```python
# call fit model function on PCA data
clusters_df_pca = fit_model(crypto_pca_df, 4)

# Display sample data
clusters_df_pca.head()
```

```python
# Create a scatter plot using hvPlot by setting 
# `x="PCA1"` and `y="PCA2"`. 
# Color the graph points with the labels found using K-Means and 
# add the crypto name in the `hover_cols` parameter to identify 
# the cryptocurrency represented by each data point.
cluster_plot_pca = clusters_df_pca.hvplot.scatter(
    x = "PCA1",
    y = "PCA2",
    by='Clusters',
    hover_cols='coin_id',
    title='Scatter Plot with Clusters using PCA (k=4)')

cluster_plot_pca
```
